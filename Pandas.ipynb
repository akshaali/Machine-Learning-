as now we want to convert all these non_rated values to NAN , so that we can use it better

```python
import numpy as np
```

```python
dtMovies[dtMovies.content_rating=='NOT RATED'].content_rating
```

```python
dtMovies[dtMovies.content_rating=='NOT RATED'].content_rating  = np.nan
```

Now here we see warning , why this warning is poping out here .
here is the reason why before this I'm gonna show something 

```python
dtMovies.content_rating.isnull().sum()
```

see, 3 if above code executed then it would be like 68 ... seems like that code didn't work . 

```python
dtMovies.loc[dtMovies.content_rating=='NOT RATED', 'content_rating']  = np.nan
```

see , no warning pops out 

```python
dtMovies.content_rating.isnull().sum()
```

reason : as above withput loc is first is setting and next is getting ...We don't know which is copy or view of the data . If it is view then it's gonna change the dataframe else doesn't . but the next approach worked now why so , Loc solves this by turning it into one operation from two operation . so , it helps and no warning pops .
